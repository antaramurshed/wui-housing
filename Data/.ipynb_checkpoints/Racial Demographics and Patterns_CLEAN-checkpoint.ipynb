## RACIAL DEMOGRPAHICS EXPLORATION

#### This notebook focuses on the census data for California on racial demographics at the county level. The dataset is from Census Reporter. There will be two maps; the one will show the non-White racial group with the highest percentage in each county  and the 2nd will highlight spatial analysis based on ethnicity.

As usual I will use labs as a guide.

```python
import pandas as pd
import geopandas as gpd
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
cacounty_race = gpd.read_file('acs2019_5yr_B03002_05000US06083.geojson')
```

```python
cacounty_race.shape
```

```python
cacounty_race.head()
```

Since this is on the county level, I only have to delete the top row which gives the data for the whole state of California.

```python
cacounty_race = cacounty_race.drop([0])
```

Now I'm going to drop a bunch of columns. I'm also going to create a few new columns later as well.

```python
list(cacounty_race)
```

```python
columns_to_keep = ['geoid',
 'name',
 'B03002001',
 'B03002002',
 'B03002003',
 'B03002004',
 'B03002005',
 'B03002006',
 'B03002007',                  
 'B03002008',
 'B03002009',
 'B03002012',
 'geometry']
```

```python
cacounty_race = cacounty_race[columns_to_keep]
```

```python
cacounty_race.head()
```

```python
list(cacounty_race)
```

```python
cacounty_race.columns = ['geoid',
 'County',
 'Total',
 'Non Hispanic',
 'Non Hispanic White',
 'Non Hispanic Black',
 'Non Hispanic American Indian and Alaska Native',
 'Non Hispanic Asian',
 'Non Hispanic Native Hawaiian and Other Pacific Islander',
 'Non Hispanic Some other race',
 'Non Hispanic Two or more races',
 'Hispanic',
 'geometry']
```

```python
# quick check before I continue
cacounty_race.head()
```

Perfect. Now, I am going to create those new columns I mentioned earlier. One new column will be named 'Non-White Not Hispanic or Latino'. To do this I will subtract data in the 'Non Hispanic White' column from the 'Non Hispanic' column. Afterwards,  I am going to convert the whole dataset's absolute numbers into percentages. Eventually, The goal is to  create maps that show the variance of non-white latinos vs latinos in each county across the state. 

```python
# new 'Non-White Not Hispanic or Latino'
cacounty_race['Non-White Not Hispanic or Latino'] = cacounty_race['Non Hispanic']-cacounty_race['Non Hispanic White']
```

```python
cacounty_race
```

Yes! Ok. Now to create percentages for all of the columns.

```python
cacounty_race['Percent Total'] = cacounty_race['Total']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic'] = cacounty_race['Non Hispanic']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic White'] = cacounty_race['Non Hispanic White']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic Black'] = cacounty_race['Non Hispanic Black']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic American Indian and Alaska Native'] = cacounty_race['Non Hispanic American Indian and Alaska Native']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic Asian'] = cacounty_race['Non Hispanic Asian']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic Native Hawaiian and Other Pacific Islander'] = cacounty_race['Non Hispanic Native Hawaiian and Other Pacific Islander']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic Some other race'] = cacounty_race['Non Hispanic Some other race']/cacounty_race['Total']*100
cacounty_race['Percent Non Hispanic Two or more races'] = cacounty_race['Non Hispanic Two or more races']/cacounty_race['Total']*100
cacounty_race['Percent Hispanic'] = cacounty_race['Hispanic']/cacounty_race['Total']*100
cacounty_race['Percent Non-White Not Hispanic or Latino'] = cacounty_race['Non-White Not Hispanic or Latino']/cacounty_race['Total']*100
```

```python
# check it
cacounty_race.head()
```

## Percentage of Latinos in Each County in the State of California

```python
fig, ax=plt.subplots(figsize=(12,12))

cacounty_race.plot(ax=ax,            
         column='Percent Hispanic',
         cmap='plasma',
         legend=True,
         scheme='user_defined',
         classification_kwds={'bins':[20,40,60,80,100]},
         edgecolor='white',
         linewidth=0.1
        )

ax.axis('off')
plt.title('Percentage of Latinos in Each County in the State of California')
```

## Percentage of Non Latinos (including whites) in Each County in the State of California

```python
fig, ax=plt.subplots(figsize=(12,12))

cacounty_race.plot(ax=ax,
         column='Percent Non Hispanic',
         cmap='plasma',
         legend=True,
         scheme='user_defined',
         classification_kwds={'bins':[20,40,60,80,100]},
         edgecolor='white',
         linewidth=0.1
        )

ax.axis('off')
plt.title('Percentage of All Non Latinos in Each County in the State of California')
```

## Percentage of White Population in Each County in the State of California

```python
fig, ax=plt.subplots(figsize=(12,12))

cacounty_race.plot(ax=ax,
         column='Percent Non Hispanic White',
         cmap='plasma',
         legend=True,
         scheme='user_defined',
         classification_kwds={'bins':[20,40,60,80,100]},
         edgecolor='white',
         linewidth=0.1
        )

ax.axis('off')
plt.title('Percentage of White Population in Each County in the State of California')
```
